```python
# create a function to construct a dictionary from existing list
def create_diction(list1,list2,list3):
    trio_price = {}
    trio_price["Open_Price"] = list1
    trio_price["High/Low_Price"] = list2
    trio_price["Close_Price"] = list3
    return trio_price
```

```python
#  create three new separate list from column Open Price, Highest Price and Closing Price of usdchf table
num3 = list(usdchf["Open Price"])
num4 = list(usdchf["Highest Price"])
num5 = list(usdchf["Closing Price"])
```

```python
shadow_price = create_diction(num3,num4,num5)
```

```python
# create a new list where each element is the price difference between Highest Price and Open/Close Price from usdchf table
pricediff1 = []
for i in range(len(shadow_price["Open_Price"])):
    if shadow_price["Open_Price"][i] < shadow_price["Close_Price"][i]:
        num1 = abs(shadow_price["High/Low_Price"][i] - shadow_price["Close_Price"][i])
    else:
        num1 = abs(shadow_price["High/Low_Price"][i] - shadow_price["Open_Price"][i])
    pricediff1.append(num1)
```

```python
# create a function to change the price difference small value by 1,000,000 times for ease of charting and save it as list
def create_list(list1):
    n = len(list1)
    new_list = []
    j = 0
    while j < n:
        new_list.append((list1[j])*1000000)
        j += 1
    return new_list
```

```python
# create a new list from column Candlestick Body of usdchf table
num7 = list(usdchf["Candlestick Body"])

# create a new list from column Upper Shadow Price of usdchf table
num8 = list(usdchf["Upper Shadow Price"])

# create a new list from column Lower Shadow Price of usdchf table
num9 = list(usdchf["Lower Shadow Price"])
```

```python
# create a set of new list that change the less than unity price difference by 1,000,000 for ease of charting
new_Candlestick = create_list(num7)
new_UpperShadow = create_list(num8)
new_LowerShadow = create_list(num9)
```

```python
# adding new columns into usdchf table
usdchf["Candlestick Body '000,000'"] = new_Candlestick
usdchf["Upper Shadow Price '000,000'"] = new_UpperShadow
usdchf["Lower Shadow Price '000,000'"] = new_LowerShadow
```
